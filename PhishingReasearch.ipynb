# Preprocessing with Scaling

```python
def preprocess_data(X):
    original_cols=X.columns
    scaler = preprocessing.Normalizer().fit(X)
    X_scaled = scaler.transform(X)
    X_scaled = pd.DataFrame(data=X_scaled, columns=original_cols)

    return X_scaled
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.neighbors import KNeighborsClassifier
```

```python
def get_scores(X_train,y_train,X_test,y_test):
    model_names = [
        "Nearest Neighbors",
        "RBF SVM",
        "Random Forest",
        "Neural Net",
        "Ada Boost"]

    classifiers = [
        KNeighborsClassifier(3),
        SVC(gamma=2, C=1),
        RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
        MLPClassifier(alpha=1, max_iter=1000),
        AdaBoostClassifier()]
    clf_scores={}
    for name, clf in zip(model_names, classifiers):
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)
        clf_scores.update({name:score})
    return clf_scores
    
    
```

# Using Feature Selection Algorithms

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from  sklearn.feature_selection import chi2
from sklearn.feature_selection import mutual_info_classif
```

```python
def get_sets_of_features(X,y,k):
    feature_selection_algos=[f_classif,chi2,mutual_info_classif]
    feature_selection_algos_names=['f_classif','chi2','mutual_info_classif']
    dict_features_by_algo={}
    for fs_algo,algo_name in zip(feature_selection_algos,feature_selection_algos_names):
        columns_algo = SelectKBest(fs_algo,k=k).fit(X, y).get_feature_names_out()
        dict_features_by_algo.update({algo_name:list(columns_algo)})
    return dict_features_by_algo

def get_results(X, y,k=10):
    X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2)
    dict_features_by_algo=get_sets_of_features(X,y,k)
    fs_algo_set_score={}
    for algo_name,features_set in dict_features_by_algo.items():
        print(len(features_set))
        X_train_sub = X_train[features_set]
        X_test_sub = X_test[features_set]

        feature_set_score=get_scores(X_train_sub,y_train,X_test_sub,y_test)
        fs_algo_set_score.update({algo_name:feature_set_score})
    return fs_algo_set_score,dict_features_by_algo
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_results_fs(results,title):
    f_classif_results = results["f_classif"]
    chi_2_results = results["chi2"]
    mutual_info_results=results["mutual_info_classif"]
    
    barWidth = 0.1
    fig = plt.subplots(figsize =(12, 8))
    # Set position of bar on X axis
    br1 = np.arange(3)
    br2 = [x + barWidth for x in br1]
    br3 = [x + barWidth for x in br2]
    br4 = [x + barWidth for x in br3]
    br5 = [x + barWidth for x in br4]

    
    KNN_algo=[results[fs_algo]["Nearest Neighbors"] for fs_algo,val in results.items() ]
    RBF_SVM=[results[fs_algo]["RBF SVM"] for fs_algo,val in results.items() ]
    RF_SVM=[results[fs_algo]["Random Forest"] for fs_algo,val in results.items() ]
    NN_NET=[results[fs_algo]["Neural Net"] for fs_algo,val in results.items() ]
    ADA_BOOST=[results[fs_algo]["Ada Boost"] for fs_algo,val in results.items() ]

    plt.bar(br1, KNN_algo, color ='yellow', width = barWidth,
        edgecolor ='grey', label ='KNN')
    plt.bar(br2, RBF_SVM, color ='g', width = barWidth,
        edgecolor ='grey', label ='RBF_SVM')
    plt.bar(br3, RF_SVM, color ='b', width = barWidth,
        edgecolor ='grey', label ='RF_SVM')
    
    plt.bar(br4, NN_NET, color ='pink', width = barWidth,
        edgecolor ='grey', label ='NN_NET')
    
    plt.bar(br5, ADA_BOOST, color ='purple', width = barWidth,
        edgecolor ='grey', label ='ADA_BOOST')
    
    plt.xlabel('Feature Selection Algorithm', fontweight ='bold', fontsize = 10)
    plt.ylabel('Accuracy', fontweight ='bold', fontsize = 10)
    plt.xticks([r + barWidth for r in range(0,3)],
            ['f_classif', 'chi2', 'mutual_info_classif'])

    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.title(title,fontdict={'family':'serif','color':'black','size':20})

    plt.show()
    
    
def plot_results_no_fs(results,title):
  
    
    results_algos=results["chi2"] 
    algo_names=["KNN","RBF SVM","RF SVM","NN NET","ADA BOOST"]
    plt.bar(*zip(*results_algos.items()))
   
    plt.xlabel('Feature Selection Algorithm', fontweight ='bold', fontsize = 10)
    plt.ylabel('Accuracy', fontweight ='bold', fontsize = 10)
    plt.xticks( rotation='vertical')

    plt.title(title,fontdict={'family':'serif','color':'black','size':20})

    plt.show()
    
    
  
```
